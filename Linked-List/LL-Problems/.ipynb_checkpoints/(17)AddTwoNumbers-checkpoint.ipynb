```python
class Node:
    def __init__(self,k):
        self.key = k
        self.next = None

def AddTwoNos(head1, head2):
    p1 = head1
    p2 = head2
    carry = 0
    sum = 0
    sumHead = None
    sumCurr = None
    while(p1 != None) or (p2 != None):
        sum = p1.key + p2.key + carry
        carry = sum//10
        if(sumHead == None):
            sumHead = Node(sum%10)
            sumCurr = sumHead
        else:
            sumCurr.next = Node(sum%10)
            sumCurr = sumCurr.next
        if(p1 != None):
            p1 = p1.next
        if(p2 != None):
            p2 = p2.next
    if(carry > 0):
        sumCurr.next = Node(carry)
    return sumHead

def printlist(head):
    curr = head
    while(curr != None):
        print(curr.key, end = " ")
        curr = curr.next

list1 = [9,9,9,9,9,9,9]
head1 = None
curr1 = None
for ele in list1:
    if(head1 == None):
        head1 = Node(ele)
        curr1 = head1
    else:
        curr1.next = Node(ele)
        curr1 = curr1.next

list2 = [9,9,9,9]
head2 = None
curr2 = None
for ele in list2:
    if(head2 == None):
        head2 = Node(ele)
        curr2 = head2
    else:
        curr2.next = Node(ele)
        curr2 = curr2.next

output = AddTwoNos(head1, head2)
printlist(output)
```
